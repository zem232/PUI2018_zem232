```python
# Manhattan Land Use count
mngrouped=Mh.groupby(['LandUse']).count().reset_index()
mngrouped
```

```python
# Brooklyn Land Use count
bkgrouped=Bk.groupby(['LandUse']).count().reset_index()
bkgrouped
```

```python
# changing just the LandUse column to integer type
mngrouped['LandUse']=mngrouped['LandUse'].astype(int)
mngrouped.head()
```

```python
bkgrouped['LandUse']=bkgrouped['LandUse'].astype(int)
bkgrouped.head()
```

```python
# merging the dataframes as new df: mnbk
mnbk=pd.merge(mngrouped,bkgrouped,on='LandUse')
mnbk.columns
```

```python
mnbk=mnbk.rename(columns={"ZipCode_x": "MN", "ZipCode_y": "BK"})
mnbk
```

```python
mnbk['LandUse']=['One & Two Family Buildings','Multi-Family Walk-Up Buildings',
                'Multi-Family Elevator Buildings','Mixed Residential & Commercial Buildings',
                'Commercial & Office Buildings','Industrial & Manufacturing','Transportation & Utility',
                'Public Facilities & Institutions','Open Space & Outdoor Recreation',
                'Parking Facilities','Vacant Land']
mnbk
```

```python
mnbk['luratio']=mnbk['MN']/mnbk['BK']
mnbk
```

```python
def ratio(df):
        if df >1:
            return 'True'
        else:
            return 'False'
```

```python
mnbk['MN>BK']=mnbk['luratio'].apply(ratio)
```

```python
mnbk
```

We discovered that Brooklyn has more counts of all Land Use types, with the exception of Multi-Family Elevator Buildings. This makes sense, since Brooklyn is a much larger borough and has around 7 times more zip codes than Manhattan. This would make sense that Manhattan has more multi-family elevator buildings, since there are much more sky-scrapers and buildings taller than 5 stories that require elevators when compared to Brooklyn.
